```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import pandas as pd
import IPython.display
import numpy as np
import soundfile as sf
%matplotlib inline
```

```python
power_of_fft2 = 14
fft_lim_n = (2**power_of_fft2) - 1
data, samplerate = sf.read("violaC5.wav", dtype='float32')
data = data.T
y, sr = (data, samplerate)
y_mono = librosa.to_mono(y)
plt.figure(figsize=(50, 40))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y_mono,n_fft=fft_lim_n)), ref=np.max)
plt.subplot(4, 2, 1)
librosa.display.specshow(D, y_axis='hz',x_axis='time',y_coords=librosa.fft_frequencies(sr=sr, n_fft=(2**power_of_fft2)))
plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')
```

```python
F = pd.DataFrame(librosa.fft_frequencies(sr=sr, n_fft=fft_lim_n))
F.to_csv('frequenciesRange.csv')
E = pd.DataFrame(D.copy())
E[int(E.shape[1]/2)].to_csv('violaDiv2.csv')
```
